```python
# byte data type is immutable
elements = [10, 20, 0, 40, 15] # this is a list of byte numbers
x = bytes(elements) # convert the list into bytes array
print(x[0])
```

```python
element=[10,266]
x=bytes(element)

print(x)
```

```python
# proof of byte data type is immutable
elements = [10, 20, 0, 40, 15]
x=(bytes(elements))
x[0]=20
print(x[0])
```

```python
# bytearray is mutable
elements = [10, 20, 0, 40, 15]
x = bytearray(elements)
x[0]=20
print(x[0])
```

```python
# list is mutable

a=[1,2,3,4]
print(a)
a[0]=6
print(a[0])
```
