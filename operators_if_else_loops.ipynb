```python
# Arithmetric Operators
print(8+9)

print(8-9)

print(8*9)

print(8/9)

print(8//9)

print(8%9)

print(8**9)

print(8-9)

print(5*6)

print(8/9)

print(9//2)

print(9%2)

print(9**2)
```

```python
# Relational Operators

print(8>9)

print(8<9)

print(8==9)

print(8!=9)

print(8>=9)

print(8<=9)
```

```python
# Logical Operators
print(1 and 0)

print(1 or 0)

print(not 1)
```

```python
# Bitwise Operators

# bitwise and
print(2 & 3)

# bitwise or
print(2 | 3)

# bitwise xor
print(2 ^ 3)

```

```python
# Membership Operators

# in/not in

print('D' not in 'Delhi')

print(1 in [2,3,4,5,6])
```
